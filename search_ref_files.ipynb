### 搜尋並將所有reference的.xrm原始檔轉成.tif

```python
import os, glob
from src.logic.data_io import load_ref
from PIL import Image

txm_fpath = 'H:\BL01B1/2025' # 原始檔的資料夾路徑
save_dir = 'G:\Background_correction_training/TXM_2025_reference' # 存.tif檔的資料夾路徑

month_dir = glob.glob(f'{txm_fpath}/*') # 搜尋底下所有月份的資料夾
print(month_dir)

for month in month_dir:
    date_dir = glob.glob(f'{month}/*')
    for date in date_dir:
        d = date.split('\\')[-1]
        for root, dirs, files in os.walk(date):
            for f in files:
                if f.endswith('.xrm') and f[:3]=='ref':
                    print(os.path.join(root, f))
                    try:
                        ref_img = load_ref(os.path.join(root, f))
                    except ValueError:
                        print('found a weird xrm file, skip it')
                        continue
                    ref_img = Image.fromarray(ref_img)
                    ref_img.save(f'{save_dir}/{d}_{f[:-4]}.tif')

print(f'total reference images: {len(os.listdir(save_dir))}')
```
